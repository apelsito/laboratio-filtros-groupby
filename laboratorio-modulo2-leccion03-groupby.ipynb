#### Ejercicio 6: ¿Cuál es el país con más títulos lanzados en el año 2020? Para resolver este ejercicio después del gropuby os puede resultar interesante el método [`idxmax`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.idxmax.html)

```python
condicion_titulitis = df["release_year"] == 2020
df_titulo = df[condicion_titulitis]
df_titulo.groupby("country")["title"].count().reset_index().sort_values(ascending=False,by="title").head(1)
```

#### Ejercicio 15: ¿Cuál es el lenguaje más frecuente en los títulos lanzados en los últimos 5 años agrupado por país?

```python
año_reciente = df["release_year"].max()
condicion_años = df["release_year"].between(año_reciente-5,año_reciente,inclusive="both")
df_ultimos_años = df[condicion_años]
df_ultimos_años.groupby(["country","Language"])["title"].count().sort_values(ascending=False).reset_index()
```
