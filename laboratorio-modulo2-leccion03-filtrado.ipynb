```python
import pandas as pd
import numpy as np
# Carga el dataset de trabajo
pd.set_option("display.max_columns", None) #Ver todo
df = pd.read_csv("datos/netflix_union.csv", index_col="Unnamed: 0")

```

#### Ejercicio 1: ¿Cuáles son todos los títulos lanzados en 2021 que son del tipo 'Movie'?

```python
condicion_lanzado = df["release_year"] == 2021
condicion_tipo = df["type"] == "Movie"
df_resultado = df[condicion_lanzado & condicion_tipo]
df_resultado.head(1)
```

```python
df_resultado["type"].unique()
```

```python
df_resultado["release_year"].unique()
```

#### Ejercicio 2: ¿Qué títulos tienen una duración de 90 minutos? Muestra solo las columnas de título y de duración.

```python
patron_regex = "^90"
condicion_duracion = df["duration"].str.contains(patron_regex,case=False,na=False)
df_duracion = df[condicion_duracion].filter(items=["title","duration"])
df_duracion.head(5)
```

```python
df_duracion["duration"].unique()
```

#### Ejercicio 3: ¿Cuáles son todos los títulos del país 'India' lanzados después de 2018?

```python
condicion_pais = df["country"] == "India"
condicion_release = df["release_year"] > 2018
df_titles_india = df[condicion_pais & condicion_release].filter(items=["title"])
df_titles_india.head(5)
```

#### Ejercicio 4: ¿Qué títulos tienen clasificación 'TV-14' o 'R'?

```python
condicion_clasificacion = ["TV-14","R"]
df[df["rating"].isin(condicion_clasificacion)].filter(items=["title"])

```

#### Ejercicio 5: ¿Cuáles son los títulos dirigidos por 'Christopher Nolan' o 'Martin Scorsese'? ¿Cuántos ª dirigido cada uno de los directores?

```python
condicion_director = ["Christopher Nolan","Martin Scorsese"]
df_directores = df[df["director"].isin(condicion_director)]#.filter(items=["title"])
df_directores.groupby("director")["title"].count().reset_index()
```

#### Ejercicio 6: ¿Qué títulos pertenecen a los países 'United States', 'United Kingdom' o 'Canada'? Dentro de estos países, ¿cuántas películas tienen más de un lenguaje?

```python
condicion_pais = ["United States","United Kingdom","Canada"]
df_titulos_pais = df[df["country"].isin(condicion_pais)]
#df_titulos_pais["Language"].unique()
patron = "/"
df_idiomas = df_titulos_pais[df_titulos_pais["Language"].str.contains(patron,case=False,na=False)]
df_idiomas.groupby("Language")["title"].count().reset_index()
```

#### Ejercicio 7: ¿Qué títulos fueron lanzados entre 2015 y 2020, inclusive?

```python
condicion_lanzados = df["release_year"].between(2015,2020,inclusive="both")
df_lanzados = df[condicion_lanzados]
df_lanzados.filter(items=["title","release_year"])
```

```python
df_lanzados.groupby("release_year")["title"].count().reset_index()
```

#### Ejercicio 8: ¿Cuáles son los títulos cuya puntuación en IMDb está entre 6.5 y 8.0 y que no sepamos cual es su duración?

```python
condicion_rating = df["IMDB Score"].between(6.5,8.0,inclusive="both")
condicion_duracion_nula = df["duration"].isnull() == True
df_rating = df[condicion_rating & condicion_duracion_nula].filter(items=["title","duration"])
df_rating
```

#### Ejercicio 9: ¿Cuáles son las películas cuya duración de 90 o 150 minutos?

```python
patron_regex1 = "^90"
patron_regex2 = "^150"
condicion_duratione = df["duration"].str.contains(patron_regex1,case=False,na=False)
condicion_duratione2 = df["duration"].str.contains(patron_regex2,case=False,na=False)
df_duracion = df[condicion_duratione | condicion_duratione2].filter(items=["title","duration"])
df_duracion
```

```python
df_duracion.groupby("duration")["title"].count().reset_index()
```

#### Ejercicio 10: ¿Cuáles son los títulos que contienen la palabra 'Love' en su título y que sean de tipo TV Show?

```python
patron_regex3 = r"\s?Love\s?"
condicion_palabra = df["title"].str.contains(patron_regex3,case=False,na=False)
condicion_tipo2 = df["type"] == "TV Show"
df_tv_love = df[condicion_palabra & condicion_tipo2]
df_tv_love
```

#### Ejercicio 11: ¿Cuáles son los títulos en los que el director tiene el apellido 'Johnson'?

```python
regex_search = r"\s?Johnson"
condicion_director_johnson = df["director"].str.contains(regex_search,case=False,na=False)
df_johnson = df[condicion_director_johnson]
df_johnson.filter(items=["title","director"])
```

#### Ejercicio 12: ¿Tenemos algún documental ('Documentaries') de Mike Tyson?

```python
df.head(1)
```

#### Ejercicio 15: Usando `query()`, filtra todos los títulos del país 'France' con una duración de 80 minutos.

#### Ejercicio 16: Filtra los títulos que son películas ('Movie') y fueron lanzados antes de 2010.

```python
condicion_peli = df["type"] == "Movie"
condicion_year = df["release_year"] < 2010
df_titulospre2010 = df[condicion_peli & condicion_year]
df_titulospre2010.filter(items=["title","release_year","type"])
```

#### Ejercicio 17: Encuentra los 10 títulos con las puntuaciones más altas en IMDb.

```python
condicion_no_nulos = df["IMDB Score"].notnull()
df_mvp = df[condicion_no_nulos]
df_mvp.nlargest(10,columns="IMDB Score").filter(items=["title","IMDB Score"])

```

#### Ejercicio 18: Encuentra los 5 títulos con las puntuaciones más bajas en IMDb.

```python
condicion_no_nulos = df["IMDB Score"].notnull()
df_mvp = df[condicion_no_nulos]
df_mvp.nsmallest(5,columns="IMDB Score").filter(items=["title","IMDB Score"])
```

#### Ejercicio 19: Filtra todos los títulos que no tienen un director asignado (valores nulos en la columna 'director').

```python
condicione_direttore = df["director"].notnull()
df[condicione_direttore].filter(items=["title","director"])
```

#### Ejercicio 20: Filtra todos los títulos que tienen información completa en las columnas 'title', 'country' y 'IMDB Score'.

```python
condicion_titulo_nonulo = df["title"].notnull()
condicion_country_nonulo = df["country"].notnull()
condicion_imdb_nonulo = df["IMDB Score"].notnull()
df_nonulos = df[condicion_titulo_nonulo & condicion_country_nonulo & condicion_imdb_nonulo]
df_nonulos.filter(items=["title","country","IMDB Score"])
```
